```python
import pandas as pd
# Importar base de dados
df = pd.read_csv('Churn_Modelling.csv')

# Visualizar base de dados
df = df.drop('RowNumber', axis=1)
df.head(10)
```

```python
import numpy as np
import seaborn as sns

myred='#CD5C5C'
myblue='#6495ED'
mygreen='#90EE90'
c = [myred, myblue, mygreen]
```

```python
# Porcentagem de clientes que sairam do banco.
df['Exited'].sum() / df['Exited'].count()
```

```python
# Análise categóricas
import matplotlib.pyplot as plt

# Sexo
df_plot = df.groupby ('Gender'). Exited.mean (). reset_index ()
df_plot
```

```python
x = df_plot['Gender']
y = df_plot['Exited']
plt.bar(x, y, color=c);
plt.title('Taxa de Churn em Relaçao ao Sexo');
plt.xlabel('Sexo');
plt.ylabel('Taxa de Churn');
```

```python
# País
df_plot = df.groupby ('Geography'). Exited.mean (). reset_index ()
df_plot
```

```python
x = df_plot['Geography']
y = df_plot['Exited']
plt.bar(x, y, color=c);
plt.title('Taxa de Churn em Relaçao ao Sexo');
plt.xlabel('Sexo');
plt.ylabel('Taxa de Churn');
```

```python
# Possuí cartao de crédito
df['HasCrCard'] = df['HasCrCard'].replace([0, 1], ['Nao', 'Sim'])
df_plot = df.groupby ('HasCrCard'). Exited.mean (). reset_index ()
df_plot
```

```python
x = df_plot['HasCrCard']
y = df_plot['Exited']
plt.bar(x, y, color=c);
plt.title('Taxa de Churn em Relaçao a Variável Cartao de Crédito');
plt.xlabel('Cartao de Crédito');
plt.ylabel('Taxa de Churn');
```

```python
# Cliente ativo
df_plot = df.groupby ('IsActiveMember'). Exited.mean (). reset_index ()
x = df_plot['IsActiveMember'].replace([0, 1], ['Nao', 'Sim'])
y = df_plot['Exited']
plt.bar(x, y, color=c);
plt.title('Taxa de Churn em Relaçao a Variável Membro Ativo');
plt.xlabel('Membro Ativo');
plt.ylabel('Taxa de Churn');
```

```python
# Box-plot Saldo Bancário pelo Churn
df['Exited_cat'] = df['Exited'].replace([0, 1], ['Nao', 'Sim'])
sns.boxplot(x='Exited_cat', y='Balance', data=df, palette=c);
plt.title('Box-plot da Variável Saldo Bancário')
plt.xlabel('Churn')
plt.ylabel('Saldo Bancário')
df.groupby(['Exited_cat'])['Balance'].mean()
```

```python
# Box-plot Idade pelo Churn
sns.boxplot(x='Exited_cat', y='Age', data=df, palette=c);
plt.title('Box-plot da Variável Idade')
plt.xlabel('Churn')
plt.ylabel('Idade')
df.groupby(['Exited'])['Age'].mean()
```

## Conclusões
- Analisando o histórico de clientes, observamos algumas características entre os clientes que cancelaram o serviço.
- Clientes da Alemanha, do sexo feminino, que não são membros ativos, com alto números de produtos, score de crédito baixo, com saldo bancário elevado e com idade entre 50 e 60 anos tem uma taxa de Churn maior que outros clientes.
    Assim clientes com essas características possuem maior risco de cancelarem o serviço bancário.

```python
# Converter classes em números
df1 = df.drop(columns = ['CustomerId', 'Surname', 'Exited_cat'])
df1
```

```python
# Dados vazios
df.isnull().sum()
```

```python
y = df1['Exited'] # Prever
X = df1 # Conj. de dados, - exited
X = df1.drop('Exited',axis = 1)
X
```

```python
# Variáveis continuas
x_cont = ['CreditScore', 'Balance', 'Age', 'EstimatedSalary', 'Tenure']
x_cont
```

```python
# Variáveis categóricas
x_cat = list(set(X)-set(x_cont))
x_cat
x_dummies = X[x_cat]
x_dummies
```

```python
# Substituindo a variável sexo para 0 e 1
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['Gender'] = le.fit_transform(X['Gender'])
X.head(10)
```

```python
# Transforma em novas variáveis numéricas
x_final = pd.get_dummies (data = X, columns=['Geography', 'NumOfProducts', 'HasCrCard'])
x_final
```

```python
# Separando em dados teste e treinamento 
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x_final, y, test_size=0.25, random_state=1)
x_treino
```

```python
# Modelo
from sklearn.tree import DecisionTreeClassifier
modelo = DecisionTreeClassifier()
modelo.fit(x_treino, y_treino)
```

```python
# Parametros para teste
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dt = DecisionTreeClassifier(random_state=61658)


params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [2,3,4,5,6,7,8,9],
}

grid = GridSearchCV(
    dt,
    params,
    cv=10,
    scoring='roc_auc',
    verbose=10,
    n_jobs=1,
)

grid.fit(x_treino, y_treino)
```

```python
# Melhor parametro
grid.best_params_
```

```python
# Melhor score
grid.best_score_
```

```python
grid.best_estimator_
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_teste, grid.predict_proba(x_teste)[:,1])
```

```python
df['proba'] = grid.predict_proba(x_final[x_treino.columns])[:,1]
df_pred = df[['CustomerId', 'proba']]
df_pred
```
